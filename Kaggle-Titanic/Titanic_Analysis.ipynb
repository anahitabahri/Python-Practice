```python
Number_Survived = len(data[data.Survived == 1])
Survival_rate = float(Number_Survived)/len(data)
print(Survival_rate)
print("Survival rate = %.2f" %Survival_rate)
```

```python
Num_Males = len(data[data.Sex == 'male'])
print("There were %d Males onboard" % Num_Males)
```

```python
Num_Females = len(data[data.Sex == 'female'])
print("There were %d Females onboard" %Num_Females)
```

```python
Males_Survived = len(data[((data.Sex == 'male') & (data.Survived == 1)) ])
Ratio_Males_Survived = float(Males_Survived)/Num_Males
print("Survival rate of Male Passengers was = %.2f" %Ratio_Males_Survived)
```

```python
Females_Survived = len(data[((data.Sex == 'female') & (data.Survived == 1)) ])
Ratio_Females_Survived = float(Females_Survived)/Num_Females
print("Survival rate of Female Passengers was = %.2f" %Ratio_Females_Survived)
```

```python
Average_age_Not_survived = data[(data.Survived == 0)].Age.mean()
print("Average age of passengers who did not survive was = %.2f" %Average_age_Not_survived)
```

```python
Classes = [1, 2, 3]
for i in Classes:
    print("Number of people in class %d was %d" %(i, len(data[data.Pclass == i])))
```

```python
Classes =[1, 2, 3]
for i in Classes:
    total_number = len(data[data.Pclass == i])
    Survived_number = len(data[(data.Pclass == i) & data.Survived == 1])
    Survival_ratio = float(Survived_number)/total_number
    print("Survival Rate for Class %i was %.2f" %(i, Survival_ratio))
```

```python
data.dropna(inplace = True)
plt.boxplot([data['Age'][data.Survived == 0], 
             data['Age'][data.Survived == 1]])
plt.xticks([1, 2], ['Not_Survived', 'Survived'])
plt.title('Survival - Age')
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.scatter(data['Fare'][(data.Pclass == 1) & (data.Survived == 1)], 
            data['Age'][(data.Pclass == 1) & (data.Survived == 1)], 
            color = 'r',
            label = 'Survived 1st class'
            )

plt.scatter(data['Fare'][(data.Pclass == 1) & (data.Survived == 0)], 
            data['Age'][(data.Pclass == 1) & (data.Survived == 0)], 
            color = 'r',
            label = 'Did not Survive 1st class',
            marker = 'x'
            )

plt.scatter(data['Fare'][(data.Pclass == 2) & (data.Survived == 1)], 
            data['Age'][(data.Pclass == 2) & (data.Survived == 1)], 
            color = 'b',
            label = 'Survived 2nd class'
            )

plt.scatter(data['Fare'][(data.Pclass == 2) & (data.Survived == 0)], 
            data['Age'][(data.Pclass == 2) & (data.Survived == 0)], 
            color = 'b',
            label = 'Did not Survive 2nd class',
            marker = 'x'
            )

plt.scatter(data['Fare'][(data.Pclass == 3) & (data.Survived == 1)], 
            data['Age'][(data.Pclass == 3) & (data.Survived == 1)], 
            color = 'g',
            label = 'Survived 3rd class'
            )

plt.scatter(data['Fare'][(data.Pclass == 3) & (data.Survived == 0)], 
            data['Age'][(data.Pclass == 3) & (data.Survived == 0)], 
            color = 'g',
            label = 'Did not Survive 3rd class',
            marker = 'x'
            )

plt.xlabel("Fare")
plt.title("Fare vs Age")
plt.legend()
plt.show()
```

```python
# specify parameters
fig = plt.figure(figsize=(18,6), dpi=1600) 
alpha=alpha_scatterplot = 0.2 
alpha_bar_chart = 0.55

ax1 = plt.subplot2grid((2,3),(0,0))
# bar graph of those who surived vs those who did not.             
data.Survived.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
ax1.set_xlim(-1, 2)
plt.title("Distribution of Survival")    

plt.subplot2grid((2,3),(0,1))
plt.scatter(data.Survived, data.Age, alpha=alpha_scatterplot)
plt.ylabel("Age")
plt.grid(b=True, which='major', axis='y')  
plt.title("Survival by Age")

ax3 = plt.subplot2grid((2,3),(0,2))
data.Pclass.value_counts().plot(kind="barh", alpha=alpha_bar_chart)
ax3.set_ylim(-1, len(data.Pclass.value_counts()))
plt.title("Class Distribution")

plt.subplot2grid((2,3),(1,0), colspan=2)
# kernel density estimate of the subset of the 1st class passenger's age
data.Age[data.Pclass == 1].plot(kind='kde')    
data.Age[data.Pclass == 2].plot(kind='kde')
data.Age[data.Pclass == 3].plot(kind='kde')
plt.xlabel("Age")    
plt.title("Age Distribution within Classes")
plt.legend(('1st Class', '2nd Class','3rd Class'),loc='best') 

ax5 = plt.subplot2grid((2,3),(1,2))
data.Embarked.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
ax5.set_xlim(-1, len(data.Embarked.value_counts()))
plt.title("Passengers per boarding location")
```

```python
plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
data.Survived.value_counts().plot(kind='barh', color="blue", alpha=.65)
ax.set_ylim(-1, len(data.Survived.value_counts())) 
plt.title("Survival Breakdown")
```

Breakdown graph by gender

```python
df_male = data.Survived[data.Sex == 'male'].value_counts().sort_index()
df_female = data.Survived[data.Sex == 'female'].value_counts().sort_index()

ax1 = fig.add_subplot(121)
df_male.plot(kind='barh',label='Male', alpha=0.55)
df_female.plot(kind='barh', color='#FA2379',label='Female', alpha=0.55)
plt.title("Survival Breakdown by Gender"); plt.legend(loc='best')
ax1.set_ylim(-1, 2) 
```

Incorporate Pclass . . . 

```python
fig = plt.figure(figsize=(18,4), dpi=1600)
alpha_level = 0.65

ax1=fig.add_subplot(141)
female_highclass = data.Survived[data.Sex == 'female'][data.Pclass != 3].value_counts()
female_highclass.plot(kind='bar', label='female, highclass', color='#FA2479', alpha=alpha_level)
ax1.set_xticklabels(["Survived", "Died"], rotation=0)
ax1.set_xlim(-1, len(female_highclass))
plt.title("Survival Breakdown: Gender and Class"); plt.legend(loc='best')

ax2=fig.add_subplot(142, sharey=ax1)
female_lowclass = data.Survived[data.Sex == 'female'][data.Pclass == 3].value_counts()
female_lowclass.plot(kind='bar', label='female, low class', color='pink', alpha=alpha_level)
ax2.set_xticklabels(["Died","Survived"], rotation=0)
ax2.set_xlim(-1, len(female_lowclass))
plt.legend(loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
male_lowclass = data.Survived[data.Sex == 'male'][data.Pclass == 3].value_counts()
male_lowclass.plot(kind='bar', label='male, low class',color='lightblue', alpha=alpha_level)
ax3.set_xticklabels(["Died","Survived"], rotation=0)
ax3.set_xlim(-1, len(male_lowclass))
plt.legend(loc='best')

ax4=fig.add_subplot(144, sharey=ax1)
male_highclass = data.Survived[data.Sex == 'male'][data.Pclass != 3].value_counts()
male_highclass.plot(kind='bar', label='male, highclass', alpha=alpha_level, color='steelblue')
ax4.set_xticklabels(["Died","Survived"], rotation=0)
ax4.set_xlim(-1, len(male_highclass))
plt.legend(loc='best')
```
